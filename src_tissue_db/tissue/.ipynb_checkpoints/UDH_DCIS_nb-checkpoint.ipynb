```python
print data_path
nuc_features = np.load(os.path.join(data_path,'all_nuc_features.npy'))
print 'shape of nuclear features is', nuc_features.shape
```

```python
im_nuc_features = [] # mean and std of all features for all cells in an image
for i in xrange(len(CaseID)):  
    indx = np.arange(start_indx[CaseID[i]-1]-1, 
                     end_indx[CaseID[i]-1])
    if len(indx) == 0:
        print CaseID[i], ' does not have cells'
    
    im_feat = nuc_features[indx,:]
    mean_feat = np.mean(im_feat, axis = 0)
    std_feat = np.std(im_feat, axis = 0)
    comb_feat = [None]*len(mean_feat)*2
    comb_feat[::2] = mean_feat
    comb_feat[1::2] = std_feat
    
    im_nuc_features.append(comb_feat)

im_nuc_features = np.vstack(im_nuc_features)
print im_nuc_features.shape
np.save(os.path.join(data_path,'im_nuc_features.npy'),im_nuc_features)
# import scipy.io
scipy.io.savemat(os.path.join(data_path,'im_nuc_features.mat'), 
                 mdict={'im_nuc_features':im_nuc_features, 'Class':Class,'ClassGrade': ClassGrade,
                        'ImageNames':CaseName, 'ImageID': CaseID})

## glm classification
ClassLabel = np.array((Class == 'DCIS').astype(int))
print ClassLabel.shape
# print Hospital == 1
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn import metrics
from sklearn.metrics import roc_curve, auc

X_train = im_nuc_features[Hospital == 1,:]
y_train = ClassLabel[Hospital == 1]
X_test = im_nuc_features[Hospital == 0,:]
y_test =  ClassLabel[Hospital == 0]
print X_train.shape, y_train.shape, X_test.shape, y_test.shape

scaler = StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)

random_state = np.random.RandomState(0)
clf = LogisticRegression(penalty='l1', warm_start=True, C = 1, random_state = random_state )
n_samples, n_features = im_nuc_features.shape

y_score = clf.fit(X_train,y_train).predict_proba(X_test)

# Compute ROC curve and ROC area for each class
fpr, tpr, _ = roc_curve(y_test, y_score[:,1])
roc_auc = auc(fpr, tpr)

print roc_auc
print fpr.shape
# print y_score
# print y_train
# Compute micro-average ROC curve and ROC area
# fpr["micro"], tpr["micro"], _ = roc_curve(y_test.ravel(), y_score.ravel())
# roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])

plt.figure()
lw = 2
plt.plot(fpr, tpr, color='darkorange',
         lw=lw, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic example')
plt.legend(loc="lower right")
plt.show()
```

```python
from itertools import cycle
from scipy import interp

scores = cross_val_score(clf, im_nuc_features, ClassLabel, cv = 5, scoring = 'roc_auc')
print scores

cv = model_selection.KFold(n_splits=5)
mean_tpr = 0.0 # mean true positive rate
mean_fpr = np.linspace(0, 1, 100)
colors = cycle(['cyan', 'indigo', 'seagreen',  'blue', 'darkorange'])
lw = 2

i = 0
for (train,test), color in zip(cv.split(im_nuc_features,ClassLabel), colors):
    probas_ = clf.fit(im_nuc_features[train], ClassLabel[train]).predict_proba(im_nuc_features[test])
    fpr, tpr, threshold = roc_curve(ClassLabel[test], probas_[:,1])
    mean_tpr += interp(mean_fpr, fpr, tpr)
    mean_tpr[0] = 0.0
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, lw=lw, color=color,
             label='ROC fold %d (area = %0.2f)' % (i, roc_auc))

    i += 1
plt.plot([0, 1], [0, 1], linestyle='--', lw=lw, color='k',
         label='Luck')

mean_tpr /= cv.get_n_splits(im_nuc_features,ClassLabel)
mean_tpr[-1] = 1.0
mean_auc = auc(mean_fpr, mean_tpr)
plt.plot(mean_fpr, mean_tpr, color='g', linestyle='--',
         label='Mean ROC (area = %0.2f)' % mean_auc, lw=lw)

plt.xlim([-0.05, 1.05])
plt.ylim([-0.05, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.show()
```

```python
# current_key = 'rwn'
# data_fname = 'train_test_list_cellclusters_features_newtraintestlabels.csv'

# current_key = 'rwc'
# data_fname = 'train_test_list_colorobjectclusters_features_newtraintestlabels.csv'

random_state = np.random.RandomState(123)
rw_ClassGrade = np.asarray(rw_ClassGrade)
rw_y_train = rw_ClassGrade[rw_Hospital == 1]
rw_y_test = rw_ClassGrade[rw_Hospital == 0]

rw_y_bin = np.asarray(label_binarize(rw_ClassGrade, classes = [0,1,2,3]))
n_classes = rw_y_bin.shape[1]
rw_y_train_bin = np.asarray(rw_y_bin[rw_Hospital == 1,:])
rw_y_test_bin = np.asarray(rw_y_bin[rw_Hospital == 0,:])
print rw_y_train_bin.shape, rw_y_test_bin.shape

colors = cycle(['cyan', 'indigo', 'seagreen',  'blue', 'darkorange', 'yellow',
                'darkblue','red','turquoise', 'pink'])

print rw_X_train.shape
clf.fit(rw_X_train,rw_y_train)
print clf.score(rw_X_train, rw_y_train)
print clf.score(rw_X_test, rw_y_test)
cv = StratifiedKFold(n_splits = 9)
# cv = model_selection.KFold(n_splits=5)
i = 0
model = []
model_perf = []

for (train,test), color in zip(cv.split(rw_X_train,rw_y_train),colors):
    clf = LogisticRegression(penalty='l1', warm_start=True, C = .1, random_state = random_state )
    # train, val 
    model.append(clf.fit(rw_X_train[train], rw_y_train[train]))
    acc = model[i].score(rw_X_train[test],rw_y_train[test])
    probas_ = model[i].predict_proba(rw_X_train[test])
    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    
    for j in range(n_classes):        
        fpr[j], tpr[j],_ = roc_curve(rw_y_train_bin[test,j], probas_[:,j])
        roc_auc[j] = auc(fpr[j], tpr[j])
    
    # micro average
    fpr["micro"], tpr["micro"],_ = roc_curve(rw_y_train_bin[test,:].ravel(), probas_.ravel())
    roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])
    
    # compute macro average
    # First aggregate all false positive rates
    all_fpr = np.unique(np.concatenate([fpr[j] for j in range(n_classes)]))
    # Then interpolate all ROC curves at this points
    mean_tpr = np.zeros_like(all_fpr)
    for j in range(n_classes):
        mean_tpr += interp(all_fpr, fpr[j], tpr[j])

    # Finally average it and compute AUC
    mean_tpr /= n_classes
    fpr["macro"] = all_fpr
    tpr["macro"] = mean_tpr
    roc_auc["macro"] = auc(fpr["macro"], tpr["macro"])      
    
    #plt.plot(fpr['macro'], tpr['macro'] , lw=lw, color=color,
    #         label='ROC fold %d (area = %0.2f)' % (i, roc_auc['macro']))
    
    print('Validation ROC fold %d (macro auc = %0.2f, micro auc = %0.2f, acc = %0.2f)' 
          % (i,roc_auc['macro'], roc_auc['micro'], acc))
    model_perf.append(acc + roc_auc['micro'] + roc_auc['macro'])
    
    ## Now onto the test model 
    acc = model[i].score(rw_X_test,rw_y_test)
    
    probas_ = model[i].predict_proba(rw_X_test)
    for j in range(n_classes):
        fpr[j], tpr[j],_ = roc_curve(rw_y_test_bin[:,j], probas_[:,j])
        roc_auc[j] = auc(fpr[j], tpr[j])
    
    # micro average
    fpr['micro'], tpr['micro'],_ = roc_curve(rw_y_test_bin.ravel(), probas_.ravel())
    roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])
    
    # compute macro average
    # First aggregate all false positive rates
    all_fpr = np.unique(np.concatenate([fpr[j] for j in range(n_classes)]))
    # Then interpolate all ROC curves at this points
    mean_tpr = np.zeros_like(all_fpr)
    for j in range(n_classes):
        mean_tpr += interp(all_fpr, fpr[j], tpr[j])
    # Finally average it and compute AUC
    mean_tpr /= n_classes
    fpr["macro"] = all_fpr
    tpr["macro"] = mean_tpr
    roc_auc["macro"] = auc(fpr["macro"], tpr["macro"])
    
    print('Test ROC fold %d, (macro auc = %0.2f, micro auc = %0.2f, acc = %0.2f)' % 
          (i, roc_auc['macro'],roc_auc['micro'],  acc))
    print 'Number of non zero coefficients  is ',np.sum(model[i].coef_ != 0), '\n'
    plt.plot(fpr['macro'], tpr['macro'] , lw=lw, color=color,
             label='ROC fold %d (area = %0.2f)' % (i, roc_auc['macro']))

    i +=1
    
plt.plot([0, 1], [0, 1], linestyle='--', lw=lw, color='k',
         label='Luck')
plt.xlim([-0.05, 1.05])
plt.ylim([-0.05, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.show()
    
# best model
best_model = np.argmax(np.asarray(model_perf))
sum_acc_4class[current_key] = model[best_model].score(rw_X_test, rw_y_test)
# micro average
probas_ = model[best_model].predict_proba(rw_X_test)
sum_fpr_micro_4class[current_key], sum_tpr_micro_4class[current_key], _ = roc_curve(
    rw_y_test_bin.ravel(), probas_.ravel())

sum_roc_auc_micro_4class[current_key] = auc(sum_fpr_micro_4class[current_key],
                                            sum_tpr_micro_4class[current_key])

for j in range(n_classes):
    fpr[j], tpr[j],_ = roc_curve(rw_y_test_bin[:,j], probas_[:,j])
    roc_auc[j] = auc(fpr[j], tpr[j])
    
# compute macro average
# First aggregate all false positive rates
all_fpr = np.unique(np.concatenate([fpr[j] for j in range(n_classes)]))
# Then interpolate all ROC curves at this points
mean_tpr = np.zeros_like(all_fpr)
for j in range(n_classes):
        mean_tpr += interp(all_fpr, fpr[j], tpr[j])
# Finally average it and compute AUC

mean_tpr /= n_classes
sum_fpr_macro_4class[current_key] = all_fpr
sum_tpr_macro_4class[current_key] = mean_tpr
sum_roc_auc_macro_4class[current_key] = auc(sum_fpr_macro_4class[current_key],
                                            sum_tpr_macro_4class[current_key])

print 'Best model %d has accuracy = %0.2f, macro roc_auc = %0.2f, micro roc_auc = %0.2f' %(
    i,sum_acc_4class[current_key],sum_roc_auc_macro_4class[current_key],sum_roc_auc_micro_4class[current_key])
```
